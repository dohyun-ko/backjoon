```python
N = int(input())

K = int(input())
apples = [list(map(int, input().split())) for _ in range(K)]
L = int(input())
raw_rotations = [input().split() for _ in range(L)]
```

```python
rotations = [(int(r[0]), r[1]) for r in raw_rotations]
```

```python
board = [[0 for x in range(N)] for y in range(N)]
# 0: empty, 1: snake, 2: apple

for apple in apples:
  n, m = apple
  board[n-1][m-1] = 2
```

```python
def change_direction(current, direction):
  if direction == 'L':
    if current == (0,1):
      return (-1, 0)
    if current == (1,0):
      return (0,1)
    if current == (0,-1):
      return (1,0)
    if current == (-1,0):
      return (0,-1)
  if direction == 'D':
    if current == (0,1):
      return (1,0)
    if current == (1,0):
      return (0,-1)
    if current == (0,-1):
      return (-1,0)
    if current == (-1,0):
      return (0,1)
  
```

```python
from queue import Queue

snake = Queue()
head = (0,0)
board[0][0] = 1
snake.put(head)

seconds = 0
direction = (0, 1)

while True:
  # print(seconds)
  # print_board()

  head = (head[0] + direction[0], head[1] + direction[1])

  seconds += 1

  if not (0 <= head[0] < N and 0 <= head[1] < N):
    break
  snake.put(head)
  if board[head[0]][head[1]] == 1:
    break

  if board[head[0]][head[1]] != 2:
    tail = snake.get()
    board[tail[0]][tail[1]] = 0

  board[head[0]][head[1]] = 1


  if rotations and rotations[0][0] == seconds:
    direction = change_direction(direction, rotations[0][1])
    rotations.pop(0)



```
